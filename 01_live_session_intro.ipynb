# 0. Welcome

**What is Python?** A free, friendly programming language used everywhere from
scientific research to web apps to AI to automating boring spreadsheets.

**What is Jupyter?** A notebook where you mix code, results, and notes
in one document. You run code in small chunks called *cells*.

**Two kinds of cells:**
- **Code cells** run Python.
- **Markdown cells** (like this one) are for notes — no code runs.

**The most important shortcut:** SHIFT + ENTER  → run the current cell.

```python
# print() is a function. It displays whatever you put inside the parentheses.
# Text goes inside quotes: "like this" or 'like this'.
print("Hello, world!")
print("Welcome to Python.")

# Anything after a # is a COMMENT. Python ignores it.
# Comments are how we leave notes for ourselves and other humans.
```

```python
# A variable is a name that points to a value. Use = to assign.
name = "Ada"
age = 28
height_m = 1.65

print(name)
print(age)
print(height_m)

# We can use them in calculations.
print("In ten years I will be:", age + 10)
```

```python
# Python has many types of data. The four you will use most:
#   string  -> text in quotes:        "Pittsburgh"
#   int     -> whole numbers:         42
#   float   -> decimal numbers:       3.14
#   bool    -> True or False (no quotes!)

a_string  = "Pittsburgh"
an_int    = 42
a_float   = 3.14
a_bool    = True

# type() tells you what kind of value something is.
print(type(a_string))
print(type(an_int))
print(type(a_float))
print(type(a_bool))
```

```python
# +  add        -  subtract
# *  multiply   /  divide
# ** power      %  remainder ("modulo")

print(5 + 3)
print(10 - 2)
print(4 * 6)
print(20 / 4)
print(2 ** 10)   # 2 to the 10th power
print(17 % 5)    # remainder of 17 divided by 5

# Save a result in a variable and reuse it.
total = 25 + 75
print("The total is", total)
```

```python
greeting = "Hello"
who      = "Ada"

# Glue strings together with +
message = greeting + ", " + who + "!"
print(message)

# An f-string (note the f before the quote) is an easier way to mix
# variables INTO text. Put each variable in {curly braces}.
fancy = f"Hello, {who}! You are {age} years old."
print(fancy)

# A few useful string tools:
sentence = "Python is awesome"
print(sentence.upper())                  # ALL CAPS
print(sentence.lower())                  # all lowercase
print(sentence.replace("awesome", "fun"))
print(len(sentence))                     # number of characters
```

```python
# A list is an ordered collection. Square brackets [ ] hold the items.
fruits = ["apple", "banana", "cherry", "date"]
print(fruits)

# Get an item by its INDEX. Python counts from 0!
print(fruits[0])    # first
print(fruits[1])    # second
print(fruits[-1])   # the LAST one (negative counts from the end)

# Add and remove items.
fruits.append("elderberry")
print(fruits)

fruits.remove("banana")
print(fruits)

# How many items?
print("There are", len(fruits), "fruits in the list.")
```

```python
# Code inside an if-block runs only when the condition is True.
# IMPORTANT: the indented lines (4 spaces) belong to the if/elif/else above.

temperature = 72

if temperature > 80:
    print("It's hot!")
elif temperature > 60:
    print("It's pleasant.")
else:
    print("Bundle up!")

# Comparisons return True or False.
print(5 > 3)         # True
print(5 == 5)        # True   (== compares; = assigns. Different!)
print("a" == "b")    # False
```

```python
# A for-loop repeats code once for each item in a collection.
for fruit in fruits:
    print(f"I like {fruit}!")

# range(n) generates 0, 1, 2, ..., n-1.
for number in range(5):
    print("Number:", number)

# Combine a loop with an if to filter what you do.
for number in range(1, 11):
    if number % 2 == 0:
        print(number, "is even")
    else:
        print(number, "is odd")
```

```python
# Goal: given a list of scores, compute the average and find how many
# scores are above it.

scores = [85, 92, 78, 90, 67, 88, 95, 73]

# sum() and len() are built-in functions. They do exactly what they sound like.
total = sum(scores)
count = len(scores)
average = total / count
print(f"Average score: {average}")

# Count how many beat the average.
above_average = 0
for score in scores:
    if score > average:
        above_average += 1   # shorthand for: above_average = above_average + 1

print(f"{above_average} of {count} scores are above average.")
```

```python
# Python has thousands of free libraries for data analysis, machine learning,
# games, web apps, automation, and more. Here is a tiny preview using
# matplotlib, the classic plotting library.

import matplotlib.pyplot as plt

months   = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
rainfall = [3.2, 2.8, 4.1, 3.9, 4.7, 3.5]

plt.bar(months, rainfall, color="steelblue")
plt.title("Monthly Rainfall (inches)")
plt.xlabel("Month")
plt.ylabel("Rainfall")
plt.show()

# In ~10 lines, we just made a publication-style chart.
# This is what Python is famous for.
```
