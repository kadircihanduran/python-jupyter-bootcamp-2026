```python
# =============================================================================
#  AFTER-SESSION NOTEBOOK 2 — DATA DETECTIVE
# =============================================================================
#  In this notebook you'll play with table-shaped data and see how Python
#  helps you ask and answer real questions about it.
#
#  You will practice:
#      - Lists and dictionaries
#      - Loops and conditionals
#      - Built-in functions like sum() and len()
#      - Writing your own functions
#      - Making your first charts
#      - A bonus peek at pandas (the most popular data library in Python)
#
#  Our (made-up) dataset: one week of sales at a small coffee shop.
#  Each row tells us the day, the drink, the price, and how many cups sold.
#
#  Each "# %%" marks a cell. Run them in order with SHIFT + ENTER.
# =============================================================================
```

```python
# A "list of dictionaries" is a great way to store table-shaped data.
#   - Each dictionary  = one row
#   - The keys         = the column names

sales = [
    {"day": "Mon", "drink": "latte",      "price": 4.50, "cups": 32},
    {"day": "Mon", "drink": "espresso",   "price": 2.75, "cups": 18},
    {"day": "Mon", "drink": "cold brew",  "price": 4.25, "cups": 25},
    {"day": "Tue", "drink": "latte",      "price": 4.50, "cups": 28},
    {"day": "Tue", "drink": "espresso",   "price": 2.75, "cups": 22},
    {"day": "Tue", "drink": "cold brew",  "price": 4.25, "cups": 30},
    {"day": "Wed", "drink": "latte",      "price": 4.50, "cups": 41},
    {"day": "Wed", "drink": "espresso",   "price": 2.75, "cups": 16},
    {"day": "Wed", "drink": "cold brew",  "price": 4.25, "cups": 35},
    {"day": "Thu", "drink": "latte",      "price": 4.50, "cups": 38},
    {"day": "Thu", "drink": "espresso",   "price": 2.75, "cups": 20},
    {"day": "Thu", "drink": "cold brew",  "price": 4.25, "cups": 40},
    {"day": "Fri", "drink": "latte",      "price": 4.50, "cups": 50},
    {"day": "Fri", "drink": "espresso",   "price": 2.75, "cups": 24},
    {"day": "Fri", "drink": "cold brew",  "price": 4.25, "cups": 55},
]

# Peek at the first row.
print(sales[0])
print(f"There are {len(sales)} rows in our dataset.")
```

```python
# Print every row in a readable way. We grab values from a dictionary
# with square brackets:  entry["day"], entry["drink"], etc.

for entry in sales:
    print(f"{entry['day']}: sold {entry['cups']} cups of {entry['drink']}")
```

```python
# We add up the "cups" value from every row.
# Start a running total at 0, then add to it inside the loop.

total_cups = 0
for entry in sales:
    total_cups += entry["cups"]    # same as: total_cups = total_cups + entry["cups"]

print(f"Total cups sold this week: {total_cups}")
```

```python
# Same pattern, but now we multiply two columns before adding.

total_revenue = 0
for entry in sales:
    total_revenue += entry["price"] * entry["cups"]

print(f"Total revenue: ${total_revenue:.2f}")
# The :.2f format spec means "show as a number with 2 decimal places."
```

```python
# Goal: a small dictionary that maps each drink name to its total cups.
# We build it up one row at a time.

cups_by_drink = {}     # start with an empty dictionary

for entry in sales:
    drink = entry["drink"]
    if drink not in cups_by_drink:
        cups_by_drink[drink] = 0          # first time we've seen this drink
    cups_by_drink[drink] += entry["cups"]

print(cups_by_drink)
```

```python
# Notice we did roughly the same thing in Step 5 — that's a clue we should
# make a function. Then we can group by "day" too without copy-pasting.

def total_cups_by(field):
    """Return a dictionary of {value: total_cups} grouped by the given field.

    Example: total_cups_by("drink")  ->  {"latte": 189, "espresso": 100, ...}
    """
    totals = {}
    for entry in sales:
        key = entry[field]
        if key not in totals:
            totals[key] = 0
        totals[key] += entry["cups"]
    return totals

print("By drink:", total_cups_by("drink"))
print("By day:  ", total_cups_by("day"))
```

```python
# Loop through the dictionary and remember the biggest value as we go.

cups_per_day = total_cups_by("day")

best_day   = None
best_count = 0
for day, cups in cups_per_day.items():
    if cups > best_count:
        best_count = cups
        best_day   = day

print(f"The busiest day was {best_day} with {best_count} cups sold.")
```

```python
import matplotlib.pyplot as plt

# Sort days into normal week order so the chart reads left to right.
day_order = ["Mon", "Tue", "Wed", "Thu", "Fri"]
counts    = [cups_per_day[d] for d in day_order]
# The line above is a "list comprehension" — a one-line for-loop that builds
# a new list. Read it as: "for each d in day_order, take cups_per_day[d]."

plt.bar(day_order, counts, color="saddlebrown")
plt.title("Cups Sold by Day")
plt.xlabel("Day of the week")
plt.ylabel("Cups sold")
plt.show()
```

```python
cups_per_drink = total_cups_by("drink")

drinks = list(cups_per_drink.keys())     # ["latte", "espresso", "cold brew"]
counts = list(cups_per_drink.values())   # [189, 100, 185]   (approximate)

plt.pie(counts, labels=drinks, autopct="%1.1f%%")
plt.title("Share of cups sold by drink")
plt.show()
```

```python
# Filter rows where more than 30 cups were sold — a "busy" row.

busy = [entry for entry in sales if entry["cups"] > 30]

print(f"Found {len(busy)} busy rows:")
for entry in busy:
    print(entry)
```

```python
# pandas is THE library for data analysis in Python. It does everything we
# did above — but in a fraction of the code.
#
# If pandas isn't installed yet, run this once in a Jupyter cell:
#       !pip install pandas
# (the leading ! tells Jupyter to run a shell command, not Python.)

import pandas as pd

df = pd.DataFrame(sales)    # turn our list of dicts into a "DataFrame"
print(df.head())            # show the first 5 rows
print()

# Total cups by drink — one line.
print(df.groupby("drink")["cups"].sum())
print()

# Add a "revenue" column and total it by day — two lines.
df["revenue"] = df["price"] * df["cups"]
print(df.groupby("day")["revenue"].sum())

# That's the door. Behind it: real datasets with millions of rows,
# CSV files, Excel files, statistical modeling, machine learning. Walk through.
```
