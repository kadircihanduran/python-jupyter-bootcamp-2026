```python
# A Mad Lib is a story with blanks. The blanks are just variables.

noun  = "dragon"
verb  = "danced"
place = "library"

story = f"The {noun} {verb} in the {place}."
print(story)

# >>> Try changing the values above and re-run. Your story changes too.
```

```python
# input() pauses the program and waits for the user to type something.
# Whatever they type is returned as a STRING.

adjective = input("Give me an adjective (a describing word): ")
animal    = input("Give me an animal: ")
food      = input("Give me a food: ")
verb_ing  = input("Give me a verb ending in -ing: ")

print()  # blank line, just for spacing
print("Here is your story:")
print(f"Once upon a time, a {adjective} {animal} was {verb_ing} in the kitchen.")
print(f"It really wanted some {food}, but the door was locked!")
```

```python
# A function is a named, reusable block of code. We "define" it with `def`,
# list its INPUTS in parentheses, and put its body indented underneath.
#
# Why bother? Because once you have a function, you can run the same logic
# many times with different inputs — without copy-pasting code.

def tell_story(adjective, animal, food, verb_ing):
    """Print a Mad Libs story using the four words provided."""
    print(f"Once upon a time, a {adjective} {animal} was {verb_ing} in the kitchen.")
    print(f"It really wanted some {food}, but the door was locked!")
    print(f"Luckily, the {animal} knew the secret password: open sesame!")

# Now we can CALL the function — as many times as we like.
tell_story("fluffy", "penguin", "pizza", "spinning")
print()
tell_story("grumpy", "elephant", "tacos", "singing")
```

```python
# `random` is a built-in Python module (a toolkit). We import it once.
# random.choice(my_list) returns one randomly-chosen item from the list.

import random

adjectives = ["sneaky", "bashful", "thunderous", "tiny", "cosmic", "salty"]
animals    = ["llama", "octopus", "wombat", "pigeon", "narwhal"]
foods      = ["pancakes", "sushi", "tacos", "ice cream", "kale"]
verbs_ing  = ["sprinting", "yodeling", "doodling", "snorting", "tap-dancing"]

random_adj    = random.choice(adjectives)
random_animal = random.choice(animals)
random_food   = random.choice(foods)
random_verb   = random.choice(verbs_ing)

print(f"Random pick: {random_adj} {random_animal} eating {random_food} while {random_verb}")
```

```python
# Generate five random stories in a row.

for round_number in range(5):
    print(f"--- Story #{round_number + 1} ---")
    tell_story(
        random.choice(adjectives),
        random.choice(animals),
        random.choice(foods),
        random.choice(verbs_ing),
    )
    print()  # blank line between stories
```

```python
# This combines input() with if/elif/else to let the user steer the story.
# .strip()  removes any extra spaces; .lower() makes "Left" and "LEFT" the same.

def adventure():
    """A tiny choose-your-own-adventure story."""
    print("You stand at the edge of a dark forest.")
    print("Two paths split in front of you.")
    choice = input("Do you go LEFT or RIGHT? ").strip().lower()

    if choice == "left":
        print("You meet a friendly fox who teaches you Python. You win!")
    elif choice == "right":
        print("You fall into a pit of marshmallows. Sticky, but survivable.")
    else:
        print("You stand still forever. The squirrels are very confused.")

# Uncomment to play:
# adventure()
```
